### Getting the CHM

```python
# dtm

with dtm_k01 as src:
    lidar_dem_im = src.read()
    sjer_ext = rio.plot.plotting_extent(src)
```

```python
# dsm 

with dsm_k01 as src:
    lidar_dsm_im = src.read()
    dsm_meta = src.profile
```

```python
# calculate canopy height model
lidar_chm = lidar_dsm_im - lidar_dem_im
```

```python
#Plot the CHM
ep.plot_bands(lidar_chm,
              cmap='viridis',
              title="Lidar Canopy Height Model (CHM)\n k01")
plt.show()
```

```python
# view histogram of the data
ep.hist(lidar_chm,
        bins=100,
        title="Distribution of Canopy Height Model Pixels")
plt.show()
```

### Crop the tif file to only that 'shape'

source (https://rasterio.readthedocs.io/en/latest/topics/masking-by-shapefile.html)


https://www.earthdatascience.org/courses/use-data-open-source-python/intro-vector-data-python/vector-data-processing/clip-vector-data-in-python-geopandas-shapely/

https://automating-gis-processes.github.io/CSC18/lessons/L6/clipping-raster.html

```python
with rio.open('/mnt/c/users/medimonster/documents/github/ant-theano-2-27/projects/data-3d-house/CHM/lidar_chm2.tiff') as src:
    out_image, out_transform = rio.mask.mask(src, [polypoly], crop=True)
```

```python
chm_k01_meta
```

```python
chm_k01_meta = src.meta.copy()
```

```python
chm_k01_meta.update({
    "driver": "GTiff",
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})
```
